```python

doc_nlp = nlp(texto)

list(doc_nlp.ents)
```

```python
# Vetor do documento, obtido pela media entre os vetores das palavras
len(doc_nlp.vector)
```

```python
import itertools as itt

# Compara todas as palavras uma a uma
# de acordo com os valores dos seus vetores
for tk1,tk2 in itt.product(doc_nlp, doc_nlp):
    similar = tk1.similarity(tk2)
    
    if similar > 0.5 and not tk1.text == tk2.text: 
        print(tk1.text, tk2.text, similar)
```

```python
from scipy import spatial

similaridade_cos = lambda a, b: 1 - spatial.distance.cosine(a,b)
```

```python
vec_presidente = doc_nlp.vocab['presidente'].vector
vec_governo = doc_nlp.vocab['governo'].vector
vec_deputado = doc_nlp.vocab['deputado'].vector
vec_camara = doc_nlp.vocab['camara'].vector

vec_jb = doc_nlp.vocab['bolsonaro'].vector
vec_rm = doc_nlp.vocab['maia'].vector
```

```python
novo_vetor = vec_governo - vec_presidente - vec_jb + vec_deputado + vec_camara
```

```python
valor_sims = []

for w in doc_nlp.vocab:
    similar = similaridade_cos(novo_vetor, w.vector)
    valor_sims.append((w, similar))
    
```

```python
valor_sims = sorted(valor_sims, key=lambda x: -x[1])
```

```python
print([w[0].text for w in valor_sims[:5]])
```

### Entidades

```python
format_ent = lambda ent: f"{ent.text} - {str(ent.start_char)} - {str(ent.end_char)} - {ent.label_} - {str(spacy.explain(ent.label_))}"

def exibe_ents(doc):
    if doc.ents:
        print("\n".join([format_ent(ent) for ent in doc.ents]))
    else:
        print("Nenhuma entidade detectada")
```

```python
exibe_ents(doc_nlp)
```
