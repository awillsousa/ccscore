```python
from bs4 import BeautifulSoup

text_html = None
ARQ_HTML='../ccscore/data/TEXT1_visl.pt.txt'
with open(ARQ_HTML, 'r') as f_html:
    text_html = "".join(f_html.readlines())
    
soup = BeautifulSoup(text_html, 'html.parser')

print(soup.text)
```

```python
class TokenVISL(object):
    def __init__(self, text, *tags):
        self.text = text
        self.tags = tags
        
    def __str__(self):
        return f"{self.text} {self.tags}"
    
```
