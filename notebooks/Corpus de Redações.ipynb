```python
import re
def get_numbers(s):
    return re.findall("([0-9]+[,.]+[0-9]+)", s)


get_numbers('1.5 zebrado">')[0]

```

```python
def ajusta_keys(x):
    if x and 'Ortografia' in x.keys():
        return {'Competência 1': float("".join(filter(str.isdigit,x['Ortografia'].replace(',','.'))))*100,
                 'Competência 2': float("".join(filter(str.isdigit,x['Adequação ao tema'].replace(',','.'))))*100,
                 'Competência 3': float("".join(filter(str.isdigit,x['Coerência'].replace(',','.'))))*100,
                 'Competência 4': float("".join(filter(str.isdigit,x['Coesão'].replace(',','.'))))*100,
                 'Competência 5': float("".join(filter(str.isdigit,x['Ética'].replace(',','.'))))*100}
    else:
        return {}
    
```

```python
    
s = [[],[],[],[],[]]
d = []
chaves = ['Ortografia', 'Adequação ao tema', 'Coerência', 'Coesão', 'Ética']
n_chaves = ['Competência 1', 'Competência 2', 'Competência 3', 'Competência 4', 'Competência 5']
for i,row in df_json_reds3.iterrows():
    novo_d = {}
    for i, k in enumerate(chaves):
        if k in row['criteria'].keys():
            v = row['criteria'][k]
            v = get_numbers(v)[0]
            v = float(v.replace(',','.'))*100
            s[i].append(v)
            novo_d[n_chaves[i]] = v
        else:
            s[i].append(None)
            novo_d[n_chaves[i]] = 0
    d.append(novo_d)

d
df_json_reds3['criteria'] = pd.Series(d)
#df_json_reds3['criteria'].map(lambda x: ajusta_keys(x))
```
